### Задача 1
Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты <br>
по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия <br>
между конверсиями в двух группах при alpha = 5%.<br>
Дайте краткие рекомендации команде . <br>

Результаты: 1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 .<br>
2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 <br>
3) Конверсии : conv1 = 1.05% conv2 = 1.47% . <br>

Нулевая гипотеза: нет статистически значимого эффекта от проведенного теста(упрощение формы заказа)
Альтернативная гипотеза: статистически значимые различия есть, можем использвать новую форму для увеличения конверсии в покупку.
Найдем доверительный интервал для разности долей:

```python
import math
p1 = 0.0105
p2 = 0.0147
n1 = 15550
n2 = 15550
interv = 1.96*math.sqrt(p1*(1-p1)/n1+p2*(1-p2)/n2)
p_low = p2-p1-interv
p_hight = p2-p1+interv
print(f"{p_low:.5},{p_hight:.5}")
```

Полученный диапазон значений не содержит 0; статистически значимые различия **обнаружены**, отвергаем нулевую гипотезу.

Используем двухстороннее тестирование, p-value меньше альфа, значит, есть статистически значимое
отличие между версиями А и В, конверсия в версии В выше, отвергаем нулевую гипотезу и 
принимаем альтернативную гипотезу о том, что **упрощение формы заказа имеет статистически значимые  отличия** от
нулевой гипотезы и **действительно увеличивает** конверсию в покупку. Рекомендую команде использовать тестовое 
изменение для всех пользователей.
